# 提案手法モデル

#### 便利な関数

```python
def plt_pic(df,file_name='test',width_low=-10,width_higt=10):
    
    plt.figure(figsize=(10,8)) # figureの縦横の大きさ
        
    width=[]
    mean=[]
    for i in range(df.shape[1]):        
        if(width_low<=(np.mean(df)[i]) and (np.mean(df)[i])<=width_higt):    
            width.append(df.columns[i])
            mean.append(np.mean(df)[i])
        
        
    for str in width:
        plt.subplot(1,2,1)
        
        plt.title("posterior_distributon")
        sns.distplot(df[str], hist=False, kde=True, rug=False, bins=10) #kdeでカーネル密度を計算
        plt.xlabel('beta')
    
    
    plt.subplot(1,2,2)
    plt.title("mean")
    pd.Series(mean,width).plot(kind='barh',grid=True)

    plt.tight_layout()
    plt.savefig('./'+file_name)
    
    plt.show()
```

```python
def pridect(beta,data):
    
    beta_mean=beta.mean()
    
    mu=0
    for i in range(beta_mean.shape[0]):
        mu+=beta_mean[i]*data[i]
    
    prob=1/(1+np.exp(-mu))
    
    
    if(prob>=0.5):
        return 1
    else:
        return 0
```

```python
def pridect(beta,data):
    
    mu=0
    for i in range(beta.shape[0]):
        mu+=beta[i]*data[i]
    
    prob=1/(1+np.exp(-mu))
    
    
    if(prob>=0.5):
        return 1
    else:
        return 0
```

```python
def score(mu_samples,X_train,y_train):
    
    beta_mean=mu_samples.mean()
    sum=0
    for i in range(X_train.shape[0]):
        if(pridect(beta_mean,X_train[i])==y_train['y'].values[i]):
            sum+=1
    
    
    return sum/y_train['y'].values.shape[0]
```

### モデル1(事前分布:正規分布)

```python
#モデルを定義する
with pm.Model() as model_1:
    
    #事前分布
    beta1=pm.Normal('beta1',mu=0,sd=3.3,shape=X_train.shape[1])
    
    #線形予測の定義 βX
    mu1=pm.math.dot(X_train,beta1)
    
    y1=pm.Bernoulli('y1',p=1/(1+pm.math.exp(-mu1)),observed=y_train)
```

```python
with model_1:
    start1=pm.find_MAP()
    step1=pm.Metropolis()
    trace1=pm.sample(30000,step=step1,start=start1,chains=1)
```

### トレース

```python
#pm.summary(trace1)
```

```python
pm.traceplot(trace1)
plt.show()
```

```python
mu_samples1 = pd.DataFrame(trace1['beta1'],columns=name.drop('y'))
mu_samples1.head()
```

### モデル3(事前分布:階層モデル　超事前分布　一様分布)

```python
#モデルを定義する
with pm.Model() as model_3:
    
    
    #超事前分布
    beta_temp_mu=pm.Normal("beta_temp_mu",mu=0,sd=3.3)
    beta_temp_sd=pm.InverseGamma("bate_temp_sd",alpha=1.0,beta=1.0)
    
    
    #事前分布
    beta3=pm.Normal('beta3',mu=beta_temp_mu,sd=beta_temp_sd,shape=X_train.shape[1])
    
    #線形予測の定義 βX
    mu3=pm.math.dot(X_train,beta3)

    y3=pm.Bernoulli('y3',p=1/(1+pm.math.exp(-mu3)),observed=y_train)
```

```python
with model_3:
    start3=pm.find_MAP()
    step3=pm.Metropolis()
    trace3=pm.sample(30000,step=step3,start=start3,chains=1)
```

### トレース

```python
pm.traceplot(trace3)
plt.show()
```
